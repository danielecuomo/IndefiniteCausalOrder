```python
from qiskit import QuantumCircuit, QuantumRegister

from qiskit.circuit import Parameter

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x1 = Parameter('$x_1$')
y1 = Parameter('$y_1$')
z1 = Parameter('$z_1$')

x2 = Parameter('$x_2$')
y2 = Parameter('$y_2$')
z2 = Parameter('$z_2$')
```

```python
#Entangled pair
e = QuantumRegister(2, name='e')

q = QuantumRegister(1, name='q')
a = QuantumRegister(2, name='a')
switch = QuantumCircuit(e, q, a, name='$\mathcal{S}$')

switch.reset(0)
switch.reset(1)
switch.reset(3)
switch.reset(4)

#Entanglement
switch.h(0)
switch.cnot(0,1)

switch.barrier()

switch.cswap(0,2,3)
switch.x(0)
switch.cswap(0,2,4)
switch.x(0)

switch.barrier()

switch.u3(x1, y1, z1, 3)
switch.u3(x2, y2, z2, 4)

switch.barrier()

switch.cswap(0,2,3)
switch.x(0)
switch.cswap(0,2,4)
switch.x(0)

switch.barrier()


switch.cswap(1,2,3)
switch.x(1)
switch.cswap(1,2,4)
switch.x(1)

switch.barrier()

switch.u3(x1, y1, z1, 4)
switch.u3(x2, y2, z2, 3)

switch.barrier()

switch.cswap(1,2,3)
switch.x(1)
switch.cswap(1,2,4)
switch.x(1)

switch.draw(output='latex', plot_barriers=False, fold=-1, scale=0.5)
```
